### *[1.1] Teste os exemplos usando Colab.*

Importante: 0 % 0, na Teoria dos Números é possível porque para qualquer q: q*0 = 0

```python
# Def da função com defesa para 0 % 0
def DIV(divisor, dividendo):
  if divisor == 0 and dividendo == 0:
        return True
  if divisor == 0 and dividendo != 0:
        return False
  return (dividendo % divisor == 0)
# Testes de fato:
print("5 | 15 :", DIV(5, 15))
print("6 | 15 :", DIV(6, 15))
print("0 | 0  :", DIV(0, 0))
print("1 | -1 :", DIV(1, -1))
```

### *[1.2] Teste estas propriedades usando Colab.*
* **i)** $c \mid b \wedge b \mid a \implies c \mid a$
* **ii)** $b \mid a \wedge b \mid a' \implies b \mid a+a'$
* **iii)** $b \mid 0 \quad \forall b \in \mathbb{Z}$
* **iv)** $1 \mid a \quad \forall a \in \mathbb{Z}$
* **v)** $0 \mid a \iff a = 0$

```python
#i) Se 'c' divide 'b' e 'b' divide 'a', então 'c' divide 'a'
print((DIV(3,6) and DIV(6,12) == DIV(3,12)))

#ii) se 'b' divide 'a' e 'b' divide 'x', então 'b' divide 'a+x'
print(DIV(3,6) and DIV(3,9) == DIV(3,6+9))

#iii) 'b divide 0 para todo número inteiro b
todos_divisiveis = True
for b in range(-1000,1000):
  if not DIV(b, 0):
    print(f"Falhou para o numero inteiro {b}")
    todos_divisiveis = False
    break
if(todos_divisiveis):
  print(f"{todos_divisiveis}: Propriedade iii) é válida para intervalo suficientemente grande")

#iv) '1' divide 'a' para todo número inteiro a
for a in range(-1000,1000):
  if not DIV(1, a):
    print(f"Falhou para o numero inteiro {a}")
    todos_divisiveis = False
    break
if(todos_divisiveis):
  print(f"{todos_divisiveis}: Propriedade iv) é válida para intervalo suficientemente grande")

#v) '0' divide 'a' se, e somente se, 'a' = 0
for a in range(-1000, 1000):
  if DIV(0, a) != (a == 0):
    print(f"falhou para o numero {a}")
    todos_divisiveis = False
    break
print(f"{todos_divisiveis}: Propriedade v) é válida para intervalo suficientemente grande")
```

### *[2.1] Escreva um teste em Python para verificar se $n$ é primo.*

```python
def eh_primo(n):
    if n < 2:
        return False
    for d in range(2, int(n**0.5) + 1):
        if n % d == 0:
            return False
    return True

for n in [2, 3, 4, 17, 48, 97, 561]:
    print(f"{n} é primo? {eh_primo(n)}")
```

### *[2.4] Use este teorema para estimar quantos números primos existem com tamanhos de 1024 bits, 1536 bits e 2048 bits* (os tamanhos usados na criptografia)

Usei pi(N) ≈ N/ln(N) com N = 2^k para cada tamanho.

```python
import math

# 2^k não cabe em float, então calculamos pelo log: pi(2^k) ≈ 2^k / (k*ln(2))
for k in [1024, 1536, 2048]:
    log10_est = k * math.log10(2) - math.log10(k * math.log(2))
    exp = int(log10_est)
    mant = 10 ** (log10_est - exp)
    print(f"{k} bits: ~ {mant:.2f}e{exp} primos")
```

### *[2.6] Construa um número primo com seu nome*

```python
import random

def fermat_primality_test(n, k=10):
    if n <= 1:
        return False
    elif n <= 3:
        return True
    elif n % 2 == 0:
        return False
    for _ in range(k):
        a = random.randint(2, n - 2)
        if pow(a, n - 1, n) != 1:
            return False
    return True

def letra_para_numero(c):
    if c == ' ':
        return '27'
    return str(ord(c.upper()) - ord('A') + 1).zfill(2)

# passo 1: frase com cerca de 45 letras
frase = "PEDRO AUGUSTO GANDRA DE ANDRADE TRABALHO DE MD"

# passo 2: converter cada letra (A=01, ..., Z=26, espaco=27) e juntar tudo
base = ''.join(letra_para_numero(c) for c in frase)

# passo 3: colocar 10 digitos aleatorios no final (terminando em 1 pra ficar impar)
sufixo = ''.join(str(random.randint(0, 9)) for _ in range(9)) + '1'
n = int(base + sufixo)

# passo 4: se nao for primo, troca os ultimos 10 digitos e testa de novo
tentativas = 0
while not fermat_primality_test(n, k=10):
    sufixo = ''.join(str(random.randint(0, 9)) for _ in range(9)) + '1'
    n = int(base + sufixo)
    tentativas += 1

print(f"frase: {frase} ({len(frase)} letras)")
print(f"achou primo em {tentativas} tentativa(s)")
print(n)
```

### *[3.1] Escreva uma função `ver_cpf()` que verifica se um CPF é válido (retornar `True`) ou não (retornar `False`)*

```python
def ver_cpf(cpf):
    cpf = ''.join(c for c in str(cpf) if c.isdigit())
    if len(cpf) != 11 or cpf == cpf[0] * 11:
        return False

    nums = [int(d) for d in cpf]

    # primeiro digito verificador
    d1 = sum((i + 1) * nums[i] for i in range(9)) % 11
    d1 = 0 if d1 == 10 else d1
    if nums[9] != d1:
        return False

    # segundo digito verificador
    d2 = sum(i * nums[i] for i in range(10)) % 11
    d2 = 0 if d2 == 10 else d2
    return nums[10] == d2

print(ver_cpf('52998224725'))
print(ver_cpf('11111111111'))
```

### *[3.2] Verifique este exemplo usando Python. Use as seguintes funções para converter entre letras e números*

```python
def numero_para_letra_v2(numero):
    letras = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    if not 0 <= numero <= 25:
        raise ValueError("O número deve estar entre 0 e 25.")
    return letras[numero]

def letra_para_numero_v2(letra):
    letras = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    if not letra.isalpha() or not letra.isupper() or len(letra) != 1:
        raise ValueError("A entrada deve ser uma única letra maiúscula (A-Z).")
    return letras.index(letra)

p = "CRYPTO"
k = "KEY"
cifrado = ""

for i, letra in enumerate(p):
    ki = k[i % len(k)]
    ci = (letra_para_numero_v2(letra) + letra_para_numero_v2(ki)) % 26
    cifrado += numero_para_letra_v2(ci)

print(f"texto claro: {p}")
print(f"chave: {k}")
print(f"cifrado: {cifrado}")
```

### *[3.6] Use Python para encontrar uma solução para*
$$
\begin{cases}
x \equiv 5 \pmod 7 \\
x \equiv 4 \pmod {11} \\
x \equiv 10 \pmod {13}
\end{cases}
$$
Verifique sua resposta.

```python
residuos = [5, 4, 10]
modulos = [7, 11, 13]
N = 1
for m in modulos:
    N *= m

x = 0
for ai, ni in zip(residuos, modulos):
    Ni = N // ni
    for ki in range(ni):
        if (Ni * ki) % ni == ai:
            x += Ni * ki
            break

x = x % N
print(f"x = {x}")

for ai, ni in zip(residuos, modulos):
    print(f"x mod {ni} = {x % ni} (esperado {ai})")
```

### *[4.1] Use Python para encontrar todos os geradores de $\mathbb{Z}_{36}^+$ e de $\mathbb{Z}_{37}^+$.*

```python
def MDC_NP(dividendo, divisor):
    while divisor != 0:
      (dividendo, divisor) = (divisor, dividendo % divisor)
    return dividendo

COPRIMOS36 = [x for x in range(1, 36) if MDC_NP(x, 36) == 1]
print(f"Coprimos de 36: {COPRIMOS36}")

COPRIMOS37 = [x for x in range(1, 37) if MDC_NP(x, 37) == 1]
print(f"Coprimos de 37: {COPRIMOS37}")
```
