Importante: 0 % 0, na Teoria dos Números é possível porque para qualquer q: q*0 = 0

```python
# Def da função com defesa para 0 % 0
def DIV(divisor, dividendo):
  if divisor == 0 and dividendo == 0:
        return True
  if divisor == 0 and dividendo != 0:
        return False
  return (dividendo % divisor == 0)
# Testes de fato:
print("5 | 15 :", DIV(5, 15))
print("6 | 15 :", DIV(6, 15))
print("0 | 0  :", DIV(0, 0))
print("1 | -1 :", DIV(1, -1))
```

### *[2.6] Construa um número primo com seu nome*

```python
import random

def fermat_primality_test(n, k=10):
    if n <= 1:
        return False
    elif n <= 3:
        return True
    elif n % 2 == 0:
        return False
    for _ in range(k):
        a = random.randint(2, n - 2)
        if pow(a, n - 1, n) != 1:
            return False
    return True

def letra_para_numero(c):
    if c == ' ':
        return '27'
    return str(ord(c.upper()) - ord('A') + 1).zfill(2)

# passo 1: frase com cerca de 45 letras
nomes_testados = ["PEDRO AUGUSTO GANDRA DE ANDRADE TRABALHO DE MD",
                    "THALES MAGALHAES PESSOA TRABALHO DE MD",
                    "MATHEUS ARAUJO PINTO CARVALHO TRABALHO DE MD"]

for frase in nomes_testados:
    # passo 2: converter cada letra (A=01, ..., Z=26, espaco=27) e juntar tudo
    base = ''.join(letra_para_numero(c) for c in frase)

    # passo 3: colocar 10 digitos aleatorios no final (terminando em 1 pra ficar impar)
    sufixo = ''.join(str(random.randint(0, 9)) for _ in range(9)) + '1'
    n = int(base + sufixo)

    # passo 4: se nao for primo, troca os ultimos 10 digitos e testa de novo
    tentativas = 0
    while not fermat_primality_test(n, k=10):
        sufixo = ''.join(str(random.randint(0, 9)) for _ in range(9)) + '1'
        n = int(base + sufixo)
        tentativas += 1

    print(f"frase: {frase} ({len(frase)} letras)")
    print(f"achou primo em {tentativas} tentativa(s)")
    print(n)
    print("-" * 50)
```

### *[3.1] Escreva uma função `ver_cpf()` que verifica se um CPF é válido (retornar `True`) ou não (retornar `False`)*

```python
def ver_cpf(cpf):
    cpf = ''.join(c for c in str(cpf) if c.isdigit())
    if len(cpf) != 11 or cpf == cpf[0] * 11:
        return False

    nums = [int(d) for d in cpf]

    # primeiro digito verificador
    d1 = sum((i + 1) * nums[i] for i in range(9)) % 11
    d1 = 0 if d1 == 10 else d1
    if nums[9] != d1:
        return False

    # segundo digito verificador
    d2 = sum(i * nums[i] for i in range(10)) % 11
    d2 = 0 if d2 == 10 else d2
    return nums[10] == d2

print(ver_cpf('52998224725'))
print(ver_cpf('11111111111'))
```

### *[4.3] Mostre que $5$ módulo $35$ não tem inverso multiplicativo.*

O inverso de $a$ módulo $N$ é um $x$ tal que $a \cdot x \equiv 1 \pmod N$. Como mdc(5, 35) = 5, o 5 não tem inverso em Z_35.

obs: Quando fazemos contas módulo  N , o número  N  é sempre o divisor. O resultado da nossa multiplicação é que será o dividendo.

```python
def inverso_modular(a, N):
  # a mod N, se encontra retorna o inverso, se não retorna 'None'
  for x in range(1, N):
    if (a * x) % N == 1:
      return x
  return None

if inverso_modular(5, 35) == None:
  print("5 mod 35 não tem inverso multiplicativo")
```

Para dois números inteiros quaisquer, $a$ e $b$, você sempre consegue encontrar dois multiplicadores inteiros ($x$ e $y$) que fazem a combinação: $$a \cdot x + b \cdot y = \text{mdc}(a, b)$$

3- Prossegue realizando divisões e subtraindo da $Lanterior$ o $quociente \cdot Latual$ para gerar a próxima linha

### **[4.8] Modifique sua implementação do algoritmo de Euclides na Parte 1 para calcular $x$ e $y$ (os coeficientes de Bézout). Use seu algoritmo para calcular $\text{mdc}(19, 1001)$ usando este método.**

**O Resultado Final:**

Quando o while terminou porque o próximo resto seria zero, as variáveis guardaram os últimos coeficientes calculados:$$ (3) \cdot 1001 + (-158) \cdot 19 = 1$$
O coeficiente $x$ (associado ao $1001$) é $3$ , já o coeficiente $y$ (associado ao $19$) é $-158$.

```python
def Coef_Bezout(dividendo, divisor):
  # Proteção para funcionar em qualquer ordem
    if divisor > dividendo:
        dividendo, divisor = divisor, dividendo

    # Guardando valores para printar ao final
    a = dividendo
    b = divisor

    x1, y1 = 1, 0 # Coeficientes da L1
    x2, y2 = 0, 1 # Coeficientes da L2

    print(f"Calculando o MDC por Bezout entre {a} e {b}:")
    print("-" * 40) # print para organização

    while divisor != 0:
      quociente = dividendo // divisor
      resto = dividendo % divisor

      # Criando os novos Coeficientes -> Lk = Lk-2 - (q*Lk-1)
      x_novo = x1 - (quociente * x2)
      y_novo = y1 - (quociente * y2)

      # Passo a Passo
      print(f"({x1})*{a} + ({y1})*{b} = {dividendo}  | Subtrair {quociente} vez(es)")

      # Atualizando os valores para a próxima divisão
      dividendo = divisor
      divisor = resto
      
      # Atualizando os valores dos coeficientes
      x1, y1 = x2, y2
      x2, y2 = x_novo, y_novo

    print("-" * 40) # print para organização
    print(f"MDC final: {dividendo}")
    print(f"Equação de Bezout: ({x1})*{a} + ({y1})*{b} = {dividendo}")

    # Retorna MDC e os coeficientes x e y 
    return dividendo, x1, y1

# Testando o exemplo:
mdc19_1001, x, y = Coef_Bezout(19, 1001)
print(f"MDC: {mdc19_1001}")
print(f"x: {x}")
print(f"y: {y}")
```

[5.1] Verifique o teorema "*Para todo inteiro positivo $n$, temos $\sum_{d|n} \varphi(d) = n$*" para $n=35,36,37$ 

```python
from math import isqrt


def phi(n):
    """Calcula o totiente de Euler."""
    if n == 1:
        return 1

    resultado = n
    x = n
    p = 2

    while p * p <= x:
        if x % p == 0:
            while x % p == 0:
                x //= p
            resultado -= resultado // p
        p += 1

    if x > 1:
        resultado -= resultado // x

    return resultado


def divisores(n):
    """Retorna lista com os divisores positivos de n."""
    divs = []

    for i in range(1, isqrt(n) + 1):
        if n % i == 0:
            divs.append(i)
            if i != n // i:
                divs.append(n // i)

    return sorted(divs)


def verificar_teorema(n):
    """Verifica Teorema."""
    divs = divisores(n)
    soma = sum(phi(d) for d in divs)

    print(f"n = {n}")
    print(f"Divisores: {divs}")
    print("Valores (d):")
    for d in divs:
        print(f"({d}) = {phi(d)}")

    print(f"Soma = {soma}")

    if soma == n:
        print("Teorema verificado.")
    else:
        print("Teorema não verificado.")

verificar_teorema(35)
print()
verificar_teorema(36)
print()
verificar_teorema(37)
```

```python
def ordem_mod(a, p):
    """Calcula a ordem multiplicativa de 'a' módulo 'p'."""
    resultado = 1
    valor = a % p

    while valor != 1:
        valor = (valor * a) % p
        resultado += 1

    return resultado


def verificar_geradores(p):
    """Identifica quais elementos são geradores."""
    geradores = []
    nao_geradores = []

    print(f"Grupo multiplicativo Z_{p}*")
    print("-" * 40)

    for a in range(1, p):
        ord_a = ordem_mod(a, p)
        print(f"Elemento {a:2d}: ordem = {ord_a}")

        if ord_a == p - 1:
            geradores.append(a)
        else:
            nao_geradores.append(a)

    print("\nGeradores:")
    print(geradores)

    print("\nNão geradores:")
    print(nao_geradores)

verificar_geradores(11)
```

*[5.3] Encontre geradores para* $p = 257, 1009, 65537$


```python
from math import isqrt


def fatores_primos(n):
    """Retorna os fatores primos distintos de n."""
    fatores = []

    d = 2
    while d * d <= n:
        if n % d == 0:
            fatores.append(d)
            while n % d == 0:
                n //= d
        d += 1

    if n > 1:
        fatores.append(n)

    return fatores


def verifica_gerador(g, p):
    """Verifica se g é um gerador."""
    ordem = p - 1
    fatores = fatores_primos(ordem)

    for q in fatores:
        if pow(g, ordem // q, p) == 1:
            return False

    return True


def listar_geradores(p):
    """Lista todos os geradores."""
    return [g for g in range(2, p) if verifica_gerador(g, p)]


for p in [257, 1009, 65537]:
    geradores = listar_geradores(p)
    print(f"p = {p}")
    print(f"Geradores: {geradores}")
    print()
```

[5.4] *Encontre um gerador para* $p=17$ *e crie uma tabela de logaritmos, de* $Z_{17}^*$ a $Z_{16}^+$.


```python
def tabela_logaritmos(p, g):
    """Gera a tabela de logaritmos discretos e retorna 'elemento -> expoente'."""
    tabela = {}
    valor = 1  # g^0

    for x in range(p - 1):
        tabela[valor] = x
        valor = (valor * g) % p

    return tabela


# Exemplo para p = 17
g = 3
tabela = tabela_logaritmos(17, g)

print(f"Gerador escolhido: g = {g}\n")
print(" h | log_g(h)")
print("---+---------")

for h in range(1, 17):
    print(f"{h:2} | {tabela[h]:2}")
```

*[5.5] Implemente uma versão simplificada do protocolo DHKE usando* $p= 2^{64} - 2^{32} + 1$ e $g=7$. *Simule os passos de Alice e Bob acima e verifique se* $k_A \equiv k_B$.


```python
import random

# Parâmetros públicos
p = 2**64 - 2**32 + 1
g = 7

# Segredos de Alice e Bob
a = random.randint(1, p - 1)
b = random.randint(1, p - 1)

# Valores públicos
A = pow(g, a, p)
B = pow(g, b, p)

# Chaves compartilhadas
k_A = pow(B, a, p)
k_B = pow(A, b, p)

# Exibição dos resultados
print(f"a = {a}")
print(f"b = {b}")
print(f"A = {A}")
print(f"B = {B}")
print(f"k_A = {k_A}")
print(f"k_B = {k_B}")

if k_A == k_B:
    print("As chaves são iguais!")
else:
    print("Erro: as chaves são diferentes.")
```
